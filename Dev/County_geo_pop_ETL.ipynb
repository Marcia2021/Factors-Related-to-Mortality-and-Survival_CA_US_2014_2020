```python
from pathlib import Path
import pandas as pd 
```

```python
#reading in urban/suburban data
df = pd.read_csv(Path('../Resources/cal_rural_urban.csv'))
df
```

```python
#renaming the columns
df.rename(columns={"county_name": "County", "urban_rural": "Description"}, inplace = True)
```

```python
df
```

```python
#dataframe for county sizes
df_sizes = pd.read_csv(Path('../Resources/county_sizes.csv'))
df_sizes
```

```python
#deleting the index column
df_sizes.drop(columns=['Unnamed: 0'], inplace= True)
```

```python
df_sizes
```

```python
#merging the county sizes and description by county name 
cal_county_info = pd.merge(df, df_sizes, how="left", on="County")
```

```python
cal_county_info.to_csv(Path('../Clean_datasets/Cal_county_geo.csv'), index = False, header = True)
```

```python
cal_county_info
```

## population data cleaning 

```python
#creating the population dataframe
df_population = pd.read_csv(Path('../Resources/Cal_county_pop.csv'))
df_population.head(5)
```
